```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split 
```

```python
plt.style.use('seaborn-white')
df = pd.read_csv("train_MNIST.csv")
```

```python
data = df.values
X = data[: , 1:]
Y = data[:, 0]
XT,Xt,YT,Yt= train_test_split(X,Y,test_size = 0.2)
```

```python
plt.figure(figsize=(10,10))
for i in range(25):
    plt.subplot(5,5,i+1)
    plt.imshow(XT[i].reshape(28,28),cmap='gray')
    plt.title(YT[i])
    plt.axis("off")
```
