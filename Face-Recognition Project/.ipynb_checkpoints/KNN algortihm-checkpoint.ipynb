```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
#to create our own dataset
from sklearn.datasets import make_blobs
X,y = make_blobs(n_samples=1000, centers=2) #we use centres=2 to create 2 classes and the values of Y are 0 and 1.
print (X.shape, y.shape)
```

```python
#plotting the given dataset Points
plt.style.use("seaborn")
plt.scatter(X[: , 0] , X[: , 1] , c  =y)
#create query point
query = np.array([-7.045,7])
plt.scatter(query[0] , query[1] , c = 'red')
plt.legend()
plt.show()
#the white is class 0 and black is class 1
```

```python
#to calculate distance from each point
def dist(x1,x2):
    return(np.sqrt(sum((x1-x2)**2)))
```

```python
#Implement KNN
def knn(X,y,query,k=5): #here K=5.. can change value for calculating more neighbours
    vals= [] #create a list to store distance and class in a tuple, i.e, vals = [(distance,class(0 or 1))]
    m = X.shape[0]
    for i in range(m):
        d = dist(query ,X[i])
        vals.append((d,y[i])) #appending the distancadn the associated class(0 or 1)
    vals = sorted(vals) #sorting the vals
    vals = vals[:k]  #to get first 'K' number of neighbours
    vals = np.array(vals) 
    new_vals = np.unique(vals[:,1], return_counts = True) #to get count of each element in vals list
    index = new_vals[1].argmax() #to get the index of the element occuring the max. no. of times
    pred = new_vals[0][index] #assigning the element(0 or 1) to the prediction vaiable
    return pred #returning the predicted class to which the query point belongs,i.e, 0 or 1
```

```python
knn(X,y,query) #can change query point from here... just add a array in place of query e.g : [-3,4] or any point you like
```
